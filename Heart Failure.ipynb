```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

**LOADING THE DATA

```python
df = pd.read_csv("C:\\Users\\gulte\\Downloads\\heart_failure_clinical_records_dataset.csv")
```

```python
df.head(100)
```

```python
df.dropna(inplace=True)
```

```python
df1 = df[["age", "creatinine_phosphokinase", "ejection_fraction","platelets", "serum_creatinine","serum_sodium", "time" ]]
```

```python
df2 = df[["anaemia", "diabetes", "high_blood_pressure", "sex","smoking","DEATH_EVENT"]]
```

```python
df3 = df.filter(df1,axis=1)
```

```python
df3.head()
```

```python
for x in df2:
    df2[x] = df2[x].astype('category')
```

```python
df2.info()
```

```python
def scalerFunction():
    for i in df1:
        scaler = MinMaxScaler(feature_range=(0,1))
        scaler.fit(np.array(df1[i]).reshape(-1,1))
        df1[i] = scaler.transform(np.array(df1[i]).reshape(-1,1))
scalerFunction()
```

```python
df1
```

```python
df1.info()
```

```python
df_new = pd.concat([df1, df2], axis=1)
```

```python
df_new
```

```python

xs = df_new.drop("DEATH_EVENT",1)
```

```python
xs = df_new.iloc[:,:-1]
ys = df_new.DEATH_EVENT
```

```python
xs
```

```python
ys
```

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy')>= 0.7):
            print("\nReached 70% accuracy so cancelling training!")
            self.model.stop_training = True

callbacks = myCallback()
```

```python
X_train, X_test, y_train, y_test = train_test_split(xs, ys, test_size = 0.6)
```

```python
model = tf.keras.models.Sequential([ 
                                    tf.keras.layers.Dense(128, activation=tf.nn.elu),
                                    tf.keras.layers.Dense(32, activation=tf.nn.elu),
                                    tf.keras.layers.Dense(16, activation=tf.nn.elu),
                                    tf.keras.layers.Dense(1, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer = 'adam',
              loss = 'categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=60,callbacks=[callbacks])
```
